# Zonas con mas desastres naturales historicamente

```
earthquakes=pd.DataFrame(df_earthquakes.COUNTRY.value_counts())
earthquakes.reset_index(level=0, inplace=True)

tsunamis=pd.DataFrame(df_tsunamis.COUNTRY.value_counts())
tsunamis.reset_index(level=0, inplace=True)

volcanos=pd.DataFrame(df_volcanos.Country.value_counts())
volcanos.reset_index(level=0, inplace=True)
volcanos['index'] = volcanos['index'].str.upper()

e_values = earthquakes['index'].unique()
t_values = tsunamis['index'].unique()
v_values = volcanos['index'].unique()

values = np.concatenate((e_values, t_values, v_values), axis=None)
countries = pd.DataFrame(values)[0].unique()
events = {}

def getCounter(df, value_counter, index_counter, value):
  row = df[df[index_counter] == value][value_counter].values
  if len(row) == 0:
    return 0
  return row[0]

for country in countries:
  t_counts = getCounter(tsunamis, 'COUNTRY', 'index', country)
  v_counts = getCounter(volcanos, 'Country', 'index', country)
  e_counts = getCounter(earthquakes, 'COUNTRY', 'index', country)
  events[country] = (t_counts + v_counts + e_counts, t_counts, v_counts, e_counts)

events = sorted(events.items(), key = lambda x : x[1], reverse=True)

from cytoolz.dicttoolz import merge

columns = ['', 'Total', 'T-sunamis', 'Volcanos', 'Earthquakes']

df_countries = pd.DataFrame(
    [merge(
        {'Country': k},
        {columns[i]: x for i, x in enumerate(v, 1)}
     ) for k, v in events]
)
df_countries
```

```
import seaborn as sns
import matplotlib.pyplot as plt

df_top = df_countries.head(10)

f, ax = plt.subplots(figsize=(29, 15))
sns.set_color_codes("pastel")
sns.barplot(x="Total", y="Country", data=df_top,
            label="Total", color="b")

# Plot the crashes where alcohol was involved
sns.set_color_codes("muted")
sns.barplot(x="T-sunamis", y="Country", data=df_top,
            label="T-sunamis", color="b")

sns.set_color_codes("colorblind")
sns.barplot(x="Earthquakes", y="Country", data=df_top,
            label="Earthquakes", color="b")

sns.set_color_codes("dark")
sns.barplot(x="Volcanos", y="Country", data=df_top,
            label="Volcanos", color="b")


ax.legend(ncol=4, loc="lower right", frameon=True)
ax.set(ylabel="",
       xlabel="Top de los 10 países con mayor número de desastres naturales historicamente")

sns.despine(left=True, bottom=True)
```

# Frecuencia de fenómenos naturales presente historicamente en japón

```
e_fre = df_earthquakes[df_earthquakes['COUNTRY'] == 'JAPAN'][['COUNTRY', 'YEAR']].groupby('YEAR').count().reset_index()
t_fre = df_tsunamis[df_tsunamis['COUNTRY'] == 'JAPAN'][['COUNTRY', 'YEAR']].groupby('YEAR').count().reset_index()
v_fre = df_volcanos[df_volcanos['Country'] == 'Japan'][['Country', 'Year']].groupby('Year').count().reset_index()
```

```
min_v = v_fre.loc[v_fre['Year'].idxmin()].Year
max_v = v_fre.loc[v_fre['Year'].idxmax()].Year

min_t = t_fre.loc[t_fre['YEAR'].idxmin()].YEAR
max_t = t_fre.loc[t_fre['YEAR'].idxmax()].YEAR

min_e = e_fre.loc[e_fre['YEAR'].idxmin()].YEAR
max_e = e_fre.loc[e_fre['YEAR'].idxmax()].YEAR
```

```
v_fre.columns = ['YEAR', 'COUNTRY']
```

```
limit_min = min(min_v, min_t, min_e)
limit_max = max(max_v, max_t, max_e)
```

```
values = t_fre.YEAR.to_list() + e_fre.YEAR.to_list() + v_fre.YEAR.to_list()

years = list(dict.fromkeys(values))

def getFreq(df, year):
  element = list(df[df.YEAR == year].COUNTRY.values)
  if len(element) == 0:
    return 0
  return element[0]

events = {}
for year in years:
  t_counts = getFreq(t_fre, year)
  v_counts = getFreq(v_fre, year)
  e_counts = getFreq(e_fre, year)
  events[year] = (t_counts, v_counts, e_counts)

columns = ['', 'T-sunamis', 'Volcanos', 'Earthquakes']

df_countries_year = pd.DataFrame(
    [merge(
        {'Year': k},
        {columns[i]: x for i, x in enumerate(events[k], 1)}
     ) for k in events]
)
```
